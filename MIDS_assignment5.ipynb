```python
import numpy as np
```

```python
points = {
    "P1": [0.1, 0.6],
    "P2": [0.15, 0.71],
    "P3": [0.08, 0.9],
    "P4": [0.16, 0.85],
    "P5": [0.2, 0.3],
    "P6": [0.25, 0.5],
    "P7": [0.24, 0.1],
    "P8": [0.3, 0.2]
}

labels = list(points.keys())
X = np.array(list(points.values()))
```

```python
m1 = np.array(points["P1"])  # Cluster C1
m2 = np.array(points["P8"])  # Cluster C2

print("Initial Centroids:")
print("m1:", m1)
print("m2:", m2)
```

```python
def distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))
```

```python
cluster1 = []
cluster2 = []

for i, point in enumerate(X):
    d1 = distance(point, m1)
    d2 = distance(point, m2)

    if d1 < d2:
        cluster1.append(labels[i])
    else:
        cluster2.append(labels[i])

print("Cluster C1:", cluster1)
print("Cluster C2:", cluster2)
```

```python
for i, label in enumerate(labels):
    if label == "P6":
        d1 = distance(X[i], m1)
        d2 = distance(X[i], m2)

        print("Distance of P6 from m1:", d1)
        print("Distance of P6 from m2:", d2)

        if d1 < d2:
            print("P6 belongs to Cluster C1")
        else:
            print("P6 belongs to Cluster C2")
```

```python
print("Population of cluster around m2 (C2):", len(cluster2))
```

```python
def compute_centroid(cluster):
    pts = [points[p] for p in cluster]
    return np.mean(pts, axis=0)

new_m1 = compute_centroid(cluster1)
new_m2 = compute_centroid(cluster2)

print("Updated m1:", new_m1)
print("Updated m2:", new_m2)
```
